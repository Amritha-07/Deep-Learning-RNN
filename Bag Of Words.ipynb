```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
sentences = []
vocab = []
for sent in data:
  x = word_tokenize(sent)
  sentence = [w.lower() for w in x if w.isalpha()]
  for word in sentence:
    if word not in vocab:
      vocab.append(word)
```

```python
vocab
```

```python
index = {}
i = 0
for word in vocab:
  index[word] = i
  i += 1
```

```python
def bag_of_words(nsent):
  vec = np.zeros(len(vocab))
  x = word_tokenize(nsent)
  sentence = [w.lower() for w in x if w.isalpha()]
  for word in sentence:
    vec[index[word]] += 1
  return vec
```

```python
print(bag_of_words("air air air pollution contamination"))
```
